1.Buatlah dataframe tersebut dengan jumlah mahasiswa 100 orang. NIM di generate secara random tetapi harus unique. NIM terdiri 3 digit. Nilai UTS di generate secara random dengan batas dari [0-100]. Konstruksi kan dataframe tersebut dari sebuah dictionary!


```python
import numpy as np
```

```python
data_dict = { 'NIM': np.random.randint(100,999,size=100),
             'NILAI': np.random.randint(0,100,size=100) } 
df = pd.DataFrame(data=data_dict) 
df
```

Dari dataframe nilai tersebut, hitung rata-rata , min-max dan quartil nya!


```python
df.describe()
```

Tampilkan mahasiswa dengan nilai >= 50!

```python
atas = df['NILAI']>50
df[atas].describe()
```

Hapus data mahasiswa dengan nim genap!

```python
df.drop(df.loc[df['NIM']%2==0].index, inplace = True)
df.head()
```

Urutkan dataframe berdasarkan NIM terkecil !

```python
df.sort_values("NIM", ascending = True)
```

Tampilkan 15 mahasiswa dengan nilai terbesar!

```python
top = df.sort_values("NILAI", ascending = False)
top.head(15)
```
